# Data Cleaning

```python
import pandas as pd
```

```python
data = pd.read_csv("Uncleaned_DS_jobs.csv")
```

```python
data
```

```python
data.shape
```

```python
data.info()
```

```python
columns_with_minus_one = data.columns[data.isin([-1, '-1']).any()].tolist()

print("Columns with -1:", columns_with_minus_one)
```

```python
for i in columns_with_minus_one:
    print(i)
    print(data[i].value_counts())
    print('''
    *
    *
    *
    *
    *
    *''')
```

```python
data['Rating'].replace(-1.0, pd.NA, inplace=True)
```

```python
data['Founded'].replace(-1, pd.NA, inplace=True)
```

```python
data.replace((-1, '-1'), pd.NA, inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Headquarters'].fillna('Unknown', inplace=True)
data['Size'].fillna('Unknown', inplace=True)
data['Type of ownership'].fillna('Unknown', inplace=True)
data['Revenue'].fillna('Unknown', inplace=True)
```

```python
data['Industry'].fillna('Unknown', inplace=True)
data['Sector'].fillna('Unknown', inplace=True)
```

```python
data['Rating'].fillna(data['Rating'].mean(), inplace=True)
```

```python
data['Founded'].fillna('Unknown', inplace=True)
```

```python
data.drop('Competitors', axis=1, inplace=True)
```

```python
data.drop(['index', 'Founded', 'Competitors'], axis=1, inplace=True)
```

```python
data
```

```python
# Assuming your DataFrame is named df
data['Salary Estimate'] = data['Salary Estimate'].str.replace(' \(Glassdoor est.\)', '')
```

```python
data['Salary Estimate']
```

```python
# Assuming your DataFrame is named df
data['Company Name'] = data['Company Name'].str.split('\n').str[0]
```

```python
data['Company Name']
```

```python
# Assuming your DataFrame is named df
data[['City', 'State']] = data['Location'].str.extract(r'([\w\s]+),\s(\w{2})')
```

```python
data
```

```python
# Assuming your DataFrame is named df
salary_extract = data['Salary Estimate'].str.extract(r'\$(\d+)K\-\$(\d+)K')

# Rename the columns
salary_extract.columns = ['Min Salary', 'Max Salary']

# Convert the values to numeric
data[['Min Salary', 'Max Salary']] = salary_extract.apply(pd.to_numeric)
```

```python
data['Avg Salary'] = (data['Min Salary'] + data['Max Salary']) / 2
```

```python
data
```

```python
# Assuming your DataFrame is named df

# Define the list of skills
skills = ['Python', 'Excel', 'Tableau', 'Power BI', 'SQL', 'AWS']

# Create columns for each skill and set initial values to 0
for skill in skills:
    data[skill] = 0

# Check and set values to 1 if the skill is found in the job description
for skill in skills:
    data[skill] = data['Job Description'].str.contains(skill, case=False).astype(int)
```

```python
data
```

```python
import pandas as pd

# Assuming your DataFrame is named data
# Create a new column 'Seniority' based on the conditions
data['Seniority'] = data['Job Description'].apply(lambda x: 'Senior' if any(seniority in x.lower() for seniority in ['sn', 'sr', 'sr.', 'senior', 'vp', 'vice president']) else pd.NA)

# Print the unique values in the 'Seniority' column to verify
print(data['Seniority'].unique())
```

```python
data
```

```python
import pandas as pd

# Assuming your DataFrame is named data
# Create a new column 'Job_Simp' based on the conditions
data['Job_Simp'] = data['Job Title'].apply(lambda x: 'Data Scientist' if 'data scientist' in x.lower() 
                                else 'Machine Learning Engineer' if 'machine learning' in x.lower() 
                                else 'Data Analyst' if 'data analyst' in x.lower() 
                                else 'Business Intelligence Analyst' if 'business intelligence' in x.lower()
                                else 'Software Engineer' if 'software engineer' in x.lower()
                                else 'Data Engineer' if 'data engineer' in x.lower()
                                else 'Other')

# Print the unique values in the 'Job_Simp' column to verify
print(data['Job_Simp'].unique())
```

# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# Assuming your DataFrame is named data
# Replace 'Rating' with the actual column name containing ratings
top_10_companies = data[['Company Name', 'Rating']].sort_values(by='Rating', ascending=False).head(10)

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x='Rating', y='Company Name', data=top_10_companies, palette='viridis')
plt.title('Top 10 Companies by Ratings')
plt.xlabel('Rating')
plt.ylabel('Company Name')
plt.show()
```

```python
# Assuming your DataFrame is named data
plt.figure(figsize=(14, 6))
sns.countplot(x='Industry', data=data, order=data['Industry'].value_counts().index, palette='muted')
plt.title('Industries with Highest Demand for Data Science Professionals')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(14, 6))
sns.countplot(x='Sector', data=data, order=data['Sector'].value_counts().index, palette='muted')
plt.title('Sectors with Highest Demand for Data Science Professionals')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import numpy as np
```

```python
# Assuming your DataFrame is named data
plt.figure(figsize=(14, 6))
sns.barplot(x='Sector', y='Avg Salary', data=data, estimator=np.mean, palette='Blues')
plt.title('Average Salary per Sector')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(x='Sector', y='Max Salary', data=data, estimator=np.max, palette='Greens')
plt.title('Maximum Salary per Sector')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Assuming your DataFrame is named data
top_5_cities = data['Location'].value_counts().head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_cities.values, y=top_5_cities.index, palette='Reds')
plt.title('Top 5 US Cities with Most Job Listings')
plt.xlabel('Number of Job Listings')
plt.ylabel('City')
plt.show()
```

```python
# Assuming your DataFrame is named data
top_industries = data.groupby('Industry')['Job Title'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 6))
sns.barplot(x=top_industries.values, y=top_industries.index, palette='Blues')
plt.title('Top 10 In-Demand Industries')
plt.xlabel('Number of Job Listings')
plt.ylabel('Industry')
plt.show()

# Average salary for each industry
avg_salary_by_industry = data.groupby('Industry')['Avg Salary'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 6))
sns.barplot(x=avg_salary_by_industry.values, y=avg_salary_by_industry.index, palette='Greens')
plt.title('Average Salary by Industry')
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.show()
```

```python
# Assuming your DataFrame is named data
plt.figure(figsize=(10, 6))
sns.barplot(x='Seniority', y='Avg Salary', data=data, estimator=np.mean, palette='BuPu')
plt.title('Average Salary by Seniority')
plt.xlabel('Seniority')
plt.ylabel('Average Salary')
plt.show()
```

```python
# Assuming your DataFrame is named data
aws_industries = data[data['AWS'] == 1]['Industry'].value_counts()

plt.figure(figsize=(14, 6))
sns.barplot(x=aws_industries.values, y=aws_industries.index, palette='Oranges')
plt.title('Industries that Use AWS')
plt.xlabel('Number of Job Listings')
plt.ylabel('Industry')
plt.show()
```

# Data Merging

```python
import pandas as pd

# Load datasets
cleaned_ds_jobs = pd.read_csv('Cleaned_DS_Jobs.csv')
custom_dataset = pd.read_csv('custom_dataset.csv')

# Merge datasets
# Left merge
left_merge = pd.merge(cleaned_ds_jobs, custom_dataset, how='left', on='Job Title')

# Right merge
right_merge = pd.merge(cleaned_ds_jobs, custom_dataset, how='right', on='Job Title')

# Outer merge
outer_merge = pd.merge(cleaned_ds_jobs, custom_dataset, how='outer', on='Job Title')

# Inner merge
inner_merge = pd.merge(cleaned_ds_jobs, custom_dataset, how='inner', on='Job Title')
```

```python

Let's discuss each type of join and answer the associated questions:

a. Left Join:
i. Explanation: A left join returns all the rows from the left (first) dataset and the matching rows from the right (second) dataset. If there is no match in the right dataset, it fills the missing values with NaN.

python
Copy code
# Example
left_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='left', on='common_key')
b. Right Join:
i. Difference from Left Join: A right join returns all the rows from the right dataset and the matching rows from the left dataset. If there is no match in the left dataset, it fills the missing values with NaN. It is essentially the opposite of a left join.
ii. Appropriate Use: It is appropriate to use a right join when you want to keep all the records from the second dataset and only the matching records from the first dataset.

python
Copy code
# Example
right_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='right', on='common_key')
c. Outer Join:
i. Usage: An outer join returns all rows when there is a match in either the left or the right dataset. If there is no match, it fills the missing values with NaN.
ii. Example: It is typically used when you want to retain all records from both datasets, and you're not concerned about missing values in either dataset.

python
Copy code
# Example
outer_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='outer', on='common_key')
d. Inner Join:
i. Suitable Scenario: An inner join returns only the rows where there is a match in both the left and the right dataset. It is suitable when you want to keep only the common records.
ii. Example: Suppose you want to analyze only the jobs that have matching entries in both datasets.

python
Copy code
# Example
inner_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='inner', on='common_key')
iii. Potential Challenges/Drawbacks: The drawback of an inner join is that you lose the information from the non-matching rows in either dataset. It may lead to a loss of potentially valuable data.

These join types provide flexibility in handling different scenarios during data merging, and the choice depends on the specific requirements of your analysis.





```
