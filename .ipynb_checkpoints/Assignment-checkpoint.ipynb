### Removing Column that are not in the cleaned dataset

```python
data.drop(['index', 'Founded', 'Competitors'], axis=1, inplace=True)
```

### Replacing Irrelevant text in the Salary Column

```python
data['Salary Estimate'] = data['Salary Estimate'].str.replace(' \(Glassdoor est.\)', '')
```

### Removing the text after /n in the Company_name column

```python
data['Company Name'] = data['Company Name'].str.split('\n').str[0]
```

### Splitting the Location column to have State and City

```python
data[['City', 'State']] = data['Location'].str.extract(r'([\w\s]+),\s(\w{2})')
```

### Replacing -1 with NA in the industry column

```python
data['Industry'].replace('-1', pd.NA, inplace=True)
```

```python
data['Industry'].values
```

### Splitting Salary Column to have Minimum, Avegrage and Maximum

```python
salary_extract = data['Salary Estimate'].str.extract(r'\$(\d+)K\-\$(\d+)K')

# Rename the columns
salary_extract.columns = ['Min Salary', 'Max Salary']

# Convert the values to numeric
data[['Min Salary', 'Max Salary']] = salary_extract.apply(pd.to_numeric)

data['Avg Salary'] = (data['Min Salary'] + data['Max Salary']) / 2
```

### Extracting the most Popular Skills((Python, excel, tableau, power_bi, sql,aws) using the Job Title Column

```python
# Define the list of skills
skills = ['Python', 'Excel', 'Tableau', 'Power BI', 'SQL', 'AWS']

# Create columns for each skill and set initial values to 0
for skill in skills:
    data[skill] = 0

# Check and set values to 1 if the skill is found in the job description
for skill in skills:
    data[skill] = data['Job Description'].str.contains(skill, case=False).astype(int)
```

### Adding Seniority to the column

```python
import pandas as pd

# Create a new column 'Seniority' based on the conditions
data['Seniority'] = data['Job Description'].apply(lambda x: 'Senior' if any(seniority in x.lower() for seniority in ['sn', 'sr', 'sr.', 'senior', 'vp', 'vice president']) else pd.NA)

# Print the unique values in the 'Seniority' column to verify
print(data['Seniority'].unique())
```

### Creating a simplified Job column for our Job Title as job_simp

```python
import pandas as pd

# Create a new column 'Job_Simp' based on the conditions
data['Job_Simp'] = data['Job Title'].apply(lambda x: 'Data Scientist' if 'data scientist' in x.lower() 
                                else 'Machine Learning Engineer' if 'machine learning' in x.lower() 
                                else 'Data Analyst' if 'data analyst' in x.lower() 
                                else 'Business Intelligence Analyst' if 'business intelligence' in x.lower()
                                else 'Software Engineer' if 'software engineer' in x.lower()
                                else 'Data Engineer' if 'data engineer' in x.lower()
                                else 'Other')

# Print the unique values in the 'Job_Simp' column to verify
print(data['Job_Simp'].unique())
```

### Determine the average and maximum salary per sector

```python
# Plot for Average Salary per Sector
plt.figure(figsize=(14, 6))
ax = sns.barplot(x='Sector', y='Avg Salary', data=data, estimator=np.mean, palette='Blues')
plt.title('Average Salary per Sector')
plt.xticks(rotation=45, ha='right')

# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()

# Plot for Maximum Salary per Sector
plt.figure(figsize=(14, 6))
ax = sns.barplot(x='Sector', y='Max Salary', data=data, estimator=np.max, palette='Greens')
plt.title('Maximum Salary per Sector')
plt.xticks(rotation=45, ha='right')

# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

## Explanation on Data Merging with different types of data

#### Left
A left join returns all the rows from the left (first) dataset and the matching rows from the right (second) dataset. If there is no match in the right dataset, it fills the missing values with NaN.

```python
left_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='left', on='common_key')
```

#### Right
A right join returns all the rows from the right dataset and the matching rows from the left dataset. If there is no match in the left dataset, it fills the missing values with NaN. It is essentially the opposite of a left join.

```python
right_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='right', on='common_key')
```

Appropriate Use: It is appropriate to use a right join when you want to keep all the records from the second dataset and only the matching records from the first dataset.

#### Outer
An outer join returns all rows when there is a match in either the left or the right dataset. If there is no match, it fills the missing values with NaN.

```python
outer_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='outer', on='common_key')
```

i. Usage: An outer join returns all rows when there is a match in either the left or the right dataset. If there is no match, it fills the missing values with NaN.


ii. Example: It is typically used when you want to retain all records from both datasets, and you're not concerned about missing values in either dataset.


#### Inner
An inner join returns only the rows where there is a match in both the left and the right dataset. It is suitable when you want to keep only the common records.

```python
inner_merge_example = pd.merge(cleaned_ds_jobs, custom_dataset, how='inner', on='common_key')
```

i. Suitable Scenario: An inner join returns only the rows where there is a match in both the left and the right dataset. It is suitable when you want to keep only the common records.


ii. Example: Suppose you want to analyze only the jobs that have matching entries in both datasets.

iii. Potential Challenges/Drawbacks: The drawback of an inner join is that you lose the information from the non-matching rows in either dataset. It may lead to a loss of potentially valuable data.
